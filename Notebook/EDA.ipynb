```python
import pandas as pd
df = pd.read_csv('Pakistan_Ecommerce_Dataset_for_EDA.csv')
df
```

```python
# Checking top 5 Categories by Orders.
print("Top 5 Categories by number of Orders:\n", df['category_name_1'].value_counts().head(5))
```

```python
# Payment Method Frequency.
print("Payment Method Distribution:\n", df['payment_method'].value_counts())

# Order Status Frequency.
print('\nOrder Status Distribution:\n', df['status'].value_counts())
```

```python
# Checking correlation of 'Payment Method' and 'Order Status'.
payment_method_corr = pd.crosstab(df['payment_method'], df['status'])
main_status = ['complete', 'canceled', 'order_refunded']
# Filtering Only needed 'Order Statuses'.
print('Correlation: Payment Method VS Order Status')
print(payment_method_corr[main_status])
```

## **Phase 4.1: Statistical Insights & Initial Findings**
### **1. Product & Market Dominance**
* **Best-Selling Categories:** **Mobiles & Tablets** is the clear market leader with **115,701 orders**, followed by **Men's Fashion**.

* **Consumer Preference:** The majority of customers in the Pakistan e-commerce landscape still rely heavily on **Cash on Delivery (COD)**, accounting for over **271,960** transactions.

### **2. Order Health & Fulfillment**
* **Status Distribution:** Out of the entire dataset, **233,658** orders reached **Complete** status, while **201,229** were **Canceled**.

* **Critical Insight:** There is a significant volume of **Canceled** and **Refunded** orders which indicates a need for deeper operational audit.

### **3. Behavioral Correlations (Payment vs. Status)**
* **The COD Paradox:** Contrary to general belief that COD has the most cancellations, it also drives the highest number of completed orders **(148,039 completions)**.

* **Digital Payment Friction:** Systems like easypay and payaxis show a high ratio of cancellations compared to completions **(e.g., easypay: 19,206 Complete vs. 52,840 Canceled)**, suggesting potential technical friction or user mistrust during the digital checkout process.

## **Phase 4.2: Temporal & Seasonality Insights**
### **1. Monthly Peak (The Sales Spike)**
* **November Dominance:** **November** is the highest performing month with **155,438** orders, likely driven by mega sales events like 11.11.

* **Consistency:** Months like **May** and **March** also show strong performance with over **60,000** orders each.

### **2. Weekly Behavioral Patterns**
* **Friday Rush:** **Friday** is the most active shopping day of the week with **129,364** orders.

* **Weekend Dip:** Interestingly, **Sunday** shows the lowest activity **(62,663 orders)**, suggesting users might prefer browsing on weekdays or **Fridays** for deals.

### **3. Category vs. Month Correlation**
* **Mobiles & Tablets in November:** A massive **31,241** orders for **Mobiles & Tablets** were recorded in **November** alone.

* **Fashion Trends:** **Men’s Fashion** peaked in **November** with **32,075** orders, showing that seasonal discounts heavily drive fashion sales.

## **Phase 4.4: The Premium Transaction Paradox (Critical Findings)**
### **1. High-Value Order Failure Rate**
* **Zero Success Rate:** Analysis of the **top 10** most expensive items **(ranging from 307,970 to 1,012,625 PKR)** shows a **0% completion rate**.

* **Cancellation Trend:** **9** out of **10** premium orders were **Canceled**, mostly involving payment methods like **Payaxis**, **Easypay**, and **Jazzvoucher**.

* **Refund Issue:** The only order that moved past cancellation worth **479,000 PKR** ended up being **Refunded**, indicating potential issues with high-ticket delivery or customer dissatisfaction.

### **2. Business Implications**
* **Security & Trust:** The high cancellation rate on expensive items (Appliances & Entertainment) suggests that either the payment gateways have strict limits for large amounts or customers are hesitant to commit to such large digital transactions.

* **Operational Leakage:** Processing these high-value "Ghost Orders" puts a strain on inventory management and operational resources without generating actual revenue.
