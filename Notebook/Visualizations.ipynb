# **Phase 5: Visualizations**

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Pakistan_Ecommerce_Dataset_for_Visualization.csv')
df.head()
```

```python
# Top 5 Categories by Order Count.
top_categories = df['category_name_1'].value_counts().head(5)

# Plotting using Pandas and Matplotlib.
ax = top_categories.plot(kind='bar', color='skyblue', edgecolor='black')
ax.bar_label(ax.containers[0], padding=3)
# Adding Titles and Labels.
plt.title('Top 5 Categories by Number of Orders')
plt.xlabel('Category Name')
plt.ylabel('Total Orders')
plt.xticks(rotation=20)
# Display the Plot.
plt.show()
```

```python
# Filtering Top 5 'Payment Methods'.
top_payments = ['cod', 'easypay', 'payaxis', 'jazzwallet', 'bankalfalah']
filtered_df = df[df['payment_method'].isin(top_payments)]
# Grouping 'Payment Method' and 'Status' Columns for Comparison.
grouped_data = filtered_df.groupby(['payment_method', 'status']).size().unstack()
# Filtering 'Status' only for 'Complete & Canceled' Orders.
final_plot_data = grouped_data[['complete', 'canceled']]
# Designing Chart.
ax = final_plot_data.plot(kind='bar', stacked=True, color=['green', 'red'], edgecolor='black')

plt.title('Payment Method: Success VS Failure')
plt.xlabel('Payment Method')
plt.ylabel('Total Orders')
plt.xticks(rotation=0)
plt.legend(title='Status')

plt.show()
```

```python
# 1. Months ka sahi order define karte hain
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

# Calculating Monthly Count and Set it into Accurate Order.
monthly_sales = df['month_name'].value_counts().reindex(month_order)

# Creating Chart.
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='orange', linewidth=2)

# Adding Labels aur Titles
plt.title('Monthly Sales Trend (Order Volume)')
plt.xlabel('Month')
plt.ylabel('Total Orders')
plt.xticks(rotation=45)
plt.grid()

# Displaying The Chart.
plt.show()
```

```python
# Fltering Top 10 Most Expensive Items. 
top_10 = df.nlargest(10, 'price')

# Creating a Horizontal Bar Chart. 
top_10.plot(kind='barh', x='category_name_1', y='price', color='red', edgecolor='black')
plt.ticklabel_format(style='plain', axis='x')

# Adding Title and Labels.
plt.title('Top 10 Most Expensive Orders (All Failed)')
plt.xlabel('Price in PKR')
plt.ylabel('Category')

# Displaying The Chart.
plt.show()
```
