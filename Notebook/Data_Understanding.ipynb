### **Data Integrity & Structure Check: Duplicates**
Before performing calculations, we must verify if the data contains redundant information or if "duplicates" actually represent valid business transactions.
#### **1. Row-Level Check**
* **Observation:** There are 464,050 duplicated rows in the dataset.

* **Analysis:** A significant portion of these duplicates are completely empty rows (Null values) that were likely added during the data extraction process.

* **Conclusion:** These rows do not represent unique transactions and must be removed in the Data Cleaning phase to ensure accurate results.
#### **2. Unnamed Columns & Noise**
* **Observation:** The dataset contains several unnecessary columns such as Unnamed: 21, Unnamed: 22, up to Unnamed: 25.

* **Impact:** These columns consume memory and provide no analytical value.

* **Action Plan:** These will be dropped during the cleaning process to streamline the dataframe.

```python
# Calculating unique values for the Grain Table
total_rows = len(df)
unique_orders = df['increment_id'].nunique()
unique_customers = df['Customer ID'].nunique()
unique_products = df['sku'].nunique()

print(f"Total Rows: {total_rows}")
print(f"Unique Orders: {unique_orders}")
print(f"Unique Customers: {unique_customers}")
print(f"Unique Products (SKUs): {unique_products}")
```

## **Phase 1: Final Report (The Transaction Structure)**
### **Understanding the Transaction Structure (Data Grain)**
Before moving to cleaning, we analyzed the unique entities within our 1,048,575 rows to understand the "Grain" of our data:
| **Entity** | **Count** | **Business Implication** |
| :---: | :---: | :---: |
| **Total Rows** | 1,048,575 | Represents every single line item recorded in the system. |
| **Unique Orders** | 408,785 | On average, each order contains ~2.5 different products. |
| **Unique Customers** | 115,326 | Indicates a healthy customer base with potential repeat purchase behavior. |
| **Unique Products** | 84,889 | Represents a diverse catalog, requiring grouping by Category for insights. |

**Analysis Insight:** Since the number of Rows is significantly higher than Unique Orders, we must treat each row as a "Line Item" rather than a single "Order." This distinction is crucial for calculating the Average Order Value (AOV) later.


```python
# Investigation of Unnamed Columns
unnamed_cols = [col for col in df.columns if 'Unnamed' in col]

for col in unnamed_cols:
    non_null_count = df[col].notnull().sum()
    unique_vals = df[col].dropna().unique()
    print(f"Column: {col}")
    print(f"- Non-Null Values: {non_null_count}")
    print(f"- Unique Values (if any): {unique_vals}")
    print("-" * 30)
```

### **Final Investigation: Unnamed Columns & Noise**
Before proceeding to data cleaning, a deep dive was conducted into the "Unnamed" columns (Unnamed: 21 to Unnamed: 25) to ensure no critical data is lost.
* **Observation:** The investigation revealed that all Unnamed columns have 0 non-null values and no unique data.

* **Result:** These columns are confirmed as "Data Noise" and will be removed in the next phase to optimize memory usage and clarity.
