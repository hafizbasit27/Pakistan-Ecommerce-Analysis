# **Phase 3: Data Cleaning**

```python
# Loading file in Python for data cleaning.
import pandas as pd
df = pd.read_csv('Pakistan_Ecommerce_Dataset_for_Cleaning.csv')
df
```

```python
# Removing null values in the item_id column of the dataset.
df = df.dropna(subset=['item_id']).copy()
df
```

```python
# Deleting all unnamed columns from the dataset.
df = df.loc[:, ~df.columns.str.contains('^Unnamed')].copy()
```

```python
df.info()
```

## **Data Cleaning & Structural Integrity**

In this phase, the raw dataset was streamlined to remove technical noise and non-functional records identified during the initial loading phase.

### **3.1 Global Data Reduction (Removing Ghost Records)**
Upon initial inspection, the dataset contained **1,048,575 rows**, but approximately **44%** of these were completely empty (Null).
* **Action:** I performed a targeted drop of all rows where **item_id** was missing. 
* **Logic:** Since **item_id** is the primary identifier for a transaction, any record without it is considered a "Ghost Row" and provides no analytical value.
* **Technical Note:** After dropping the rows, I created a deep copy of the dataframe using **.copy()** to avoid memory fragmentation and **SettingWithCopyWarnings** in future steps.
* **Outcome:** The dataset was reduced to **584,524 rows**, which represents the actual transactional data.

### **3.2 Removing Irrelevant Metadata (Unnamed Columns)**
The original file contained 5 "Unnamed" columns (labeled 21 to 24/25) that carried no headers and zero data.
* **Action:** These columns were dropped to optimize memory usage and focus only on relevant business variables.
* **Current Shape:** The dataset now stands at **584,524 rows and 21 columns**.

### **3.3 Roadmap: Individual Column Audit (Post-Mortem)**
To ensure the highest data quality, I am now transitioning from "Global Cleaning" to "Granular Auditing." Each column will be inspected individually for:
1. **Data Type Correction:** Ensuring IDs are integers, prices are floats, and dates are in proper Datetime format.
2. **Value Standardization:** Fixing inconsistent labels in categorical columns like **status**.
3. **Logic Validation:** Ensuring numerical values (price, quantity) are logical and positive.

```python
# Converting data types of ID columns (item_id & customerID) to integer data type.
# Converting 'item_id' to integer data type.
df['item_id'] = df['item_id'].astype('int64')

# Checking if there any missing values in 'Customer ID' column and then converting it to integer data type.
df['Customer ID'] = df['Customer ID'].fillna(0).astype('int64')

# Verification
print(df[['item_id', 'Customer ID']].dtypes)
print(df[['item_id', 'Customer ID']])
```

```python
# 1.Cleaning and standardizing 'Status' column.
df['status'] = df['status'].replace(['\\N', 'nan'], 'unknown')
df['status'] = df['status'].fillna('unknown').str.lower()

# 2.Cleaning and standardizing 'SKU' column.
df['sku'] = df['sku'].fillna('unknown').str.lower()

# 3. Cleaning and standardizing 'Payment Method' column.
df['payment_method'] = df['payment_method'].fillna('unknown').str.lower()

# 4. Checking 'null' in 'category_name_1' column.
print(f"Nulls in Category: {df['category_name_1'].isnull().sum()}")
# Removing 'null', if any exists.
df['category_name_1'] = df['category_name_1'].fillna('unknown').str.lower()

# Verification of 'SKU' and 'Payment Method' columns.
print(f"Nulls in SKU: {df['sku'].isnull().sum()}")
print(f"Nulls in Payment Method: {df['payment_method'].isnull().sum()}")
print("\nFirst 5 SKUs:", df['sku'].head())
```

### **3.4 Granular Data Transformation (Post-Mortem Audit)**

After restructuring the dataframe, I performed an individual column audit to align data types with business logic and handle hidden inconsistencies.

#### **A. Identifier Standardization**
* **Columns:** **item_id**, **Customer ID**
* **Action:** Converted from **float64** to **int64**. 
* **Reasoning:** Transactional and Customer IDs are discrete whole numbers. Removing decimals ensures indexing integrity and reduces the memory footprint.

#### **B. Categorical Consistency & Missing Value Imputation**
* **Columns:** **status**, **sku**, **payment_method**, **category_name_1**
* **Action:** 
    1. Identified technical artifacts **(\\N, nan)** and actual nulls.
    2. Imputed all missing entries with a generic **'unknown'** label to maintain dataset completeness.
    3. Converted all strings to **lowercase** for uniform grouping.
* **Reasoning:** SKU and Status are critical for product-level analysis. Standardization prevents data fragmentation and ensures 100% record availability for categorical aggregation.

```python
# 1. Converting 'created_at' and 'Working Date' to proper datetime
df['created_at'] = pd.to_datetime(df['created_at'])
df['Working Date'] = pd.to_datetime(df['Working Date'])
# Creating 'day_name' & 'month_name' columns for EDA.
df['day_name'] = df['created_at'].dt.day_name()
df['month_name'] = df['created_at'].dt.month_name()

# 2. Converting 'Year' and 'Month' to integer.
df['Year'] = df['Year'].astype('int64')
df['Month'] = df['Month'].astype('int64')

# Verification
print(df[['created_at', 'Working Date', 'Year', 'Month']].dtypes)
print(df[['created_at', 'day_name', 'month_name']].head())
print("\nSample Dates:\n", df[['created_at', 'Working Date', 'Year', 'Month']].head())
```

#### **C. Feature Engineering for Time-Series Analysis**
* **Columns Created:** **day_name**, **month_name**
* **Action:** Extracted categorical time components from the **created_at** datetime object.
* **Reasoning:** To address key business questions regarding peak sales periods. These features allow us to visualize order frequency by day of the week and identify seasonal trends across different months.

```python
# 1. Total rows before dropping.
print(f"Rows before dropping: {len(df)}")

# 2. Dropping rows where 'Grand Total' or 'Discount Amount' is negative.
df = df[df['grand_total'] >= 0].copy()
df = df[df['discount_amount'] >= 0].copy()

# 3. Converting 'Quantity Orders' to integer.
df['qty_ordered'] = df['qty_ordered'].astype('int64')

# 4. Final verification
print(f"Rows after dropping negative values: {len(df)}")
print(f"Total rows removed: 584524 - {len(df)}")

# 5. Check dtypes one last time
print("\nFinal Financial Dtypes:")
print(df[['price', 'qty_ordered', 'grand_total', 'discount_amount']].dtypes)
```

#### **D. Quantitative Integrity & Outlier Treatment**
* **Columns:** **price**, **qty_ordered**, **grand_total**, **discount_amount**
* **Action:** 
    1. Identified and **dropped 76 rows** with negative **grand_total**.
    2. Identified and **dropped 3 rows** with negative **discount_amount**.
    3. Converted **qty_ordered** to **int64** for logical consistency.
* **Reasoning:** Financial metrics cannot be negative in a standard sales context. Dropping these records ensures that aggregate calculations (Sum, Mean) reflect actual business revenue without being skewed by data entry errors or logical outliers.

### **3.5 Final Data Cleaning & Structural Integrity Log**

The data cleaning phase is now concluded. Every column has been audited for technical artifacts, logical consistency, and optimal data typing.
